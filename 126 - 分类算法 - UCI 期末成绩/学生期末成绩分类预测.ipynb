```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
data = pd.read_csv('DATA.csv', sep=';')
target = data['GRADE']
```

```python
data
```

```python
# 各个特征值对结果的相关性（皮卡尔相关系数-正相关性）
arr_pear = []
for x in data.columns[1:-1]:
    r = pearsonr(x=data.loc[:, x], y=target)
    arr_pear.append(r)
    print(r)
```

```python
index = ['学生年龄', '性别', '高中毕业类型', '奖学金类型', '附加工作', '定期的艺术或体育活动', '您是否有伴侣', '总工资（如果有）', '到大学的交通', '塞浦路斯住宿类型', '母亲的教育', '父亲的教育', '姐妹/兄弟人数（如果有）', '父母身份', '母亲的职业', '父亲职业', '每周学习时间', '阅读频率（非科学书籍/期刊）', '阅读频率（科学书籍/期刊）', '参加与部门相关的研讨会/会议', '您的项目/活动对您成功的影响', '出席上课', '准备期中考试1', '准备期中考试2', '准备期中考试', '课堂听力', '讨论提高了我在课程中的兴趣和成功', '翻转课堂', '累积成绩上学期平均分', '毕业时预期累积平均绩点', '课程 ID']
arr_pear = pd.DataFrame(data=arr_pear, index=index)
arr_pear.sort_values(by=1, inplace=True)
arr_pear
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(x=arr_pear.index, y=arr_pear.iloc[:, 0], color='b', label='皮卡尔逊正相关性')
plt.scatter(x=arr_pear.index, y=arr_pear.iloc[:, 1], marker='+', color='r', label='皮卡尔逊负相关性')
plt.legend()
plt.title('各种特征值对最终考试成绩的影响（皮卡尔相关系数）\n高等教育学生成绩评估', fontsize=20)
plt.xticks(rotation=70)
plt.xlabel('影响因素', fontsize=20)
plt.ylabel('皮卡尔逊相关系数', fontsize=20)
plt.grid(linestyle='--', alpha=0.5)
plt.savefig('./result_pear.png')
plt.show()
```
