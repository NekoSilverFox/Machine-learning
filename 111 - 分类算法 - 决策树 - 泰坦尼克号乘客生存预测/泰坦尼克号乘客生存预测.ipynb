# 泰坦尼克号乘客生存预测
获取数据集

```python
import pandas as pd
data = pd.read_csv(filepath_or_buffer='./titanic.csv')
data
```

---

截取数据集，只保留影响可能比较大的

```python
data_x = data.loc[:, ['pclass', 'age', 'sex']]
data_y = data.loc[:, 'survived']
```

```python
data_x
```

```python
data_y
```

---

处理 age 中的 NaN 值

```python
data_x.fillna(value=data_x['age'].mean(), inplace=True)
```

```python
data_x
```

---

转换为字典

```python
data_x = data_x.to_dict(orient='records')
```

```python
data_x
```

```python
# 划分数据集
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(data_x, data_y, random_state=233)
```

```python
# 使用字典特征提取，提取
from sklearn import feature_extraction
transfer = feature_extraction.DictVectorizer()
x_train = transfer.fit_transform(X=x_train)
x_test = transfer.transform(X=x_test)
```

```python
# 导入决策树预估器
# 决策树不需要算距离，所以不需要进行无量纲化
from sklearn import tree
estimator = tree.DecisionTreeClassifier(criterion='entropy', max_depth=None)
estimator.fit(X=x_train, y=y_train)

print('准确率为：', estimator.score(X=x_test, y=y_test))
```

```python
# 进行决策树的可视化
# 【重点】因为在字典特征提取后（ont-hot 编码），一个特征可能变成了多个列。所以在 feature_names 那可以使用 字典特征提取的 transfer 进行特征传入
tree.export_graphviz(decision_tree=estimator, out_file='tree_titanic.dot', feature_names=transfer.feature_names_) 
```
