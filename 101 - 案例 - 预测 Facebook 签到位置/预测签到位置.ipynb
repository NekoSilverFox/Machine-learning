```python
import pandas as pd

# 导入数据
data_train = pd.read_csv(filepath_or_buffer='./FBlocation/train.csv')
```

```python
data_train
```

```python
# 因为数据量过多，所以减少数据。只取位置在 x < 9, y < 9 之间的
data_train = data_train.query('x < 9 & y < 9')
```

```python
# 将时间转化为 日、星期几的形式
time = pd.to_datetime(data_train['time'], unit='s')  # 时间戳转 time 类型
time = pd.DatetimeIndex(time) # 转换为可通过 . 索引的数据类型

# 因为数据集里全是 1970 年 1 月，所以只取日、时、星期几
# 将他们添加为新列
data_train['day'] = time.day
data_train['hour'] = time.hour
data_train['weekday'] = time.weekday
```

```python
data_train
```

```python
# 筛选掉签到次数过少的位置（小于 10 次）
count_place = data_train.groupby(data_train['place_id']).count()
count_place
```

```python
# 只取其中一列就好
count_place = count_place.iloc[:, 1]
data_train = data_train[data_train['place_id'].isin(count_place.index.values)]  # 【重点】布尔索引
```

```python
data_train
```

```python
# 确定特征值和目标值
x_train = data_train[['x', 'y', 'accuracy', 'day', 'hour', 'weekday']]  # 【重点】注意这里是两个括号
y_train = data_train['place_id']

# 数据集划分
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x_train, y_train, random_state=233)
```

```python
x_train
```

```python
x_test
```

```python
# 特征预处理（标准化）
from sklearn import preprocessing

# 实例化转换器对象
transfer = preprocessing.StandardScaler()

# 调用 fit 和 transform
x_train = transfer.fit_transform(X=x_train)
x_test = transfer.transform(X=x_test)
```

```python
# 机器学习 KNN + CV
# 实例化估计器对象
from sklearn import neighbors
estimator = neighbors.KNeighborsClassifier()

# 导入CV转换器对象
param_grid = {'n_neighbors': [i for i in range(2, 10)]}  # 【重点】KNN 算法参数，注意是要以字典的形式
estimator = model_selection.GridSearchCV(estimator=estimator, param_grid=param_grid, cv=10)

# 模型训练
estimator.fit(X=x_train, y=y_train)
```

```python
# 模型评估：
score = estimator.score(X=x_test, y=y_test)
print("预测的准确率为:\n", score)

y_predict =estimator.predict(x_test)
print("预测值和真实值的对比情况:\n", y_predict == y_test)

print('交叉验证中最高的准确率：\n', estimator.best_score_)
print('网格搜索中最好的参数：\n', estimator.best_params_)
print('每次交叉验证后的验证集准确率结果和训练集准确率结果：\n', estimator.cv_results_)
print('最好的估计器：\n', estimator.best_estimator_)
```
